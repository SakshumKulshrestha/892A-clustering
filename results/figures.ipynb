```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_upgma = pd.read_csv('upgma_result.csv')
df_usearch = pd.read_csv('usearch_result.csv')
df_vsearch = pd.read_csv('vsearch_result.csv')

df_upgma['run_time'] = df_upgma['run_time'].apply(lambda x: float(x[0]))
df_upgma['max-mem'] = df_upgma['max-mem'].apply(lambda x: float(x[:2]))
df_upgma['perc_cluster_singleton'] = df_upgma['perc_cluster_singleton'].apply(lambda x: float(x[:-1]))
df_upgma['clust_tool'] = 'upgma'
```

```python
df_usearch['run_time'] = df_usearch['run_time'].apply(lambda x: float(x[0]))
df_vsearch['run_time'] = df_vsearch['run_time'].apply(lambda x: float(x[3:]))
```

```python
df_usearch['max-mem'] = df_usearch['max-mem'].apply(lambda x: float(x[:-2]))
df_vsearch['max-mem'] = df_vsearch['max-mem'].apply(lambda x: float(x[:-2]))
df_usearch['perc_cluster_singleton'] = df_usearch['perc_cluster_singleton'].apply(lambda x: float(x[:-1]))
df_vsearch['perc_cluster_singleton'] = df_vsearch['perc_cluster_singleton'].apply(lambda x: float(x[:-1]))
```

## Similarity vs runtime analysis

```python
get_sim = lambda x, y: x.loc[(x['dataset_name'] == 'Youssef') & (x['clust_tool'] == y)]
usearch = get_sim(df_usearch, 'usearch')
vsearch = get_sim(df_vsearch, 'vsearch')
upgma = get_sim(df_upgma, 'upgma')

plt.plot(usearch['id'], usearch['run_time'], label="usearch")
plt.plot(vsearch['id'], np.log(vsearch['run_time']), label="vsearch")
plt.plot(upgma['id'], np.log(upgma['run_time']), label="natural log upgma")
plt.legend(loc='upper left')
plt.xlabel("Percent Similarity Per Cluster")
plt.ylabel("Run Time (secs)")
plt.savefig('sim_vs_runtime_youssef.png')
plt.show()
```

## seq length vs runtime

```python
get_sim = lambda x, y: x.loc[(x['id'] == 0.950) & (x['clust_tool'] == y)][-7:]
usearch = get_sim(df_usearch, 'usearch')
vsearch = get_sim(df_vsearch, 'vsearch')
upgma = get_sim(df_upgma, 'upgma')

plt.scatter(usearch['avg_nt'], usearch['run_time'], label="usearch")
plt.scatter(vsearch['avg_nt'], np.log(vsearch['run_time']), label="natural log vsearch")
plt.scatter(upgma['avg_nt'], np.log10(upgma['run_time']), label="natural log upgma")
plt.legend(loc='upper left')
plt.xlabel("Sequence Length")
plt.ylabel("Run Time (secs)")
plt.savefig('seq_length_vs_runtime.png')
plt.show()
```

## Seq length vs max-mem

```python
get_sim = lambda x, y: x.loc[(x['id'] == 0.950) & (x['clust_tool'] == y)][-7:]
usearch = get_sim(df_usearch, 'usearch')
vsearch = get_sim(df_vsearch, 'vsearch')
upgma = get_sim(df_upgma, 'upgma')

plt.scatter(usearch['avg_nt'], usearch['max-mem'], label="usearch")
plt.scatter(vsearch['avg_nt'], vsearch['max-mem'], label="vsearch")
plt.scatter(upgma['avg_nt'], upgma['max-mem'], label="log upgma")
plt.legend(loc='upper left')
plt.xlabel("Sequence Length")
plt.ylabel("Max Memory (Mbs)")
plt.savefig('seq_length_vs_memory.png')
plt.show()
```

## number of singleton clusters 

```python
get_sim = lambda x, y: x.loc[(x['id'] == 0.950) & (x['clust_tool'] == y)]
usearch = get_sim(df_usearch, 'usearch')
vsearch = get_sim(df_vsearch, 'vsearch')
upgma = get_sim(df_upgma, 'upgma')
total = pd.concat([usearch, vsearch, upgma])[['dataset_name', 'clust_tool', 'perc_cluster_singleton']]
total = total.pivot(index='dataset_name', columns='clust_tool')['perc_cluster_singleton']
total.reset_index(inplace=True)
total.plot(x="dataset_name", y=["upgma", "vsearch", "usearch"], kind="bar")
plt.xlabel("Dataset")
plt.ylabel("Percent of Clusters Singleton")
plt.savefig('num_singleton_clusters.png')
plt.show()
```

```python
get_sim = lambda x, y: x.loc[(x['id'] == 0.950) & (x['clust_tool'] == y)]
usearch = get_sim(df_usearch, 'usearch')
vsearch = get_sim(df_vsearch, 'vsearch')
upgma = get_sim(df_upgma, 'upgma')
total = pd.concat([usearch, vsearch, upgma])[['dataset_name', 'clust_tool', 'avg_clust_size']]
total = total.pivot(index='dataset_name', columns='clust_tool')['avg_clust_size']
total.reset_index(inplace=True)
total = total.loc[total['dataset_name'] != 'QuinceArtficialGSFLXClean']
total.plot(x="dataset_name", y=["upgma", "vsearch", "usearch"], kind="bar")
plt.xlabel("Dataset")
plt.ylabel("Percent of Clusters Singleton")
plt.savefig('avg_cluster_size.png')
plt.show()
```
